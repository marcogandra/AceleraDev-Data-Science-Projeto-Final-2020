# Pre-Processamento

```python
csv_path = '../Data/'

market = pd.read_csv(f'{csv_path}estaticos_market.csv', index_col='Unnamed: 0')
port1 = pd.read_csv(f'{csv_path}estaticos_portfolio1.csv', index_col='Unnamed: 0')
port2 = pd.read_csv(f'{csv_path}estaticos_portfolio2.csv', index_col='Unnamed: 0')
port3 = pd.read_csv(f'{csv_path}estaticos_portfolio3.csv', index_col='Unnamed: 0')

ids = market['id']

port2 = port2.merge(market,on='id')
port3 = port3.merge(market,on='id')
```

### Dropando features redundantes

```python
features_redundantes = ['de_ramo', 'de_natureza_juridica','idade_emp_cat',
                     'nm_divisao','nm_micro_regiao','de_faixa_faturamento_estimado',
                     'de_faixa_faturamento_estimado_grupo','dt_situacao','de_saude_rescencia',
                       'idade_maxima_socios','idade_minima_socios']

market.drop(features_redundantes, 1, inplace = True)
```

### DataFrame Auxiliar

```python
aux = pd.DataFrame({'tipo': market.dtypes,
                    'missing': market.isna().sum(),
                    'size':market.shape[0],
                    'unicos': market.nunique()},index = market.columns)

aux['percentual'] = round(1 - (aux['missing'] / aux['size']),2)
```

```python
aux.sort_values('percentual',ascending=False)
```

### Tratamento de valores missing:

```python
market['qt_coligados'].fillna(0, inplace=True)
market['qt_socios'].fillna(1, inplace=True)
market['qt_socios_pf'].fillna(1, inplace=True)
market['qt_socios_pj'].fillna(0, inplace=True)
market['qt_funcionarios'].fillna(0, inplace=True)
market['tx_crescimento_12meses'].fillna(0, inplace=True)
market['tx_crescimento_24meses'].fillna(0, inplace=True)
market['fl_optante_simei'].fillna('False',inplace=True)
market['fl_optante_simples'].fillna('False',inplace=True)
market['nm_meso_regiao'].fillna('OUTROS',inplace=True)
market['nu_meses_rescencia'].fillna(market['nu_meses_rescencia'].median(),inplace=True)
market['vl_faturamento_estimado_aux'].fillna(market['vl_faturamento_estimado_aux'].median(),inplace=True)
market['vl_faturamento_estimado_grupo_aux'].fillna(market['vl_faturamento_estimado_aux'].median(),inplace=True) # Substituir pela mediana do faturamento estimado, pois provavelmente a empresa n√£o faz parte de um grupo
market.loc[market['sg_uf_matriz'].isna(), 'sg_uf_matriz'] = market.loc[market['sg_uf_matriz'].isna(), 'sg_uf']
market['de_nivel_atividade'].fillna('MUITO BAIXA',inplace=True)
market['de_saude_tributaria'].fillna('VERMELHO',inplace=True)
```

```python
market['idade_media_socios'].fillna(market['idade_media_socios'].median(), inplace=True)
market['empsetorcensitariofaixarendapopulacao'].fillna(market['empsetorcensitariofaixarendapopulacao'].median(), inplace=True)
```

### Feature Engineering

```python
market['porc_st_regular'] = market['qt_socios_st_regular']/market['qt_socios']
```

```python
market['socio_pep'] = 0
market.loc[market['qt_socios_pep'] > 0, 'socio_pep'] = 1
```

```python
market['coligada_exterior'] = 0
market.loc[market['qt_coligados_exterior'] > 0, 'coligada_exterior'] = 1
```

```python
market['porc_socios_pf'] = market['qt_socios_pf']/market['qt_socios']
market['porc_socios_pj'] = market['qt_socios_pj']/market['qt_socios']
```

### Tratamento de Outliers

```python
market = market[market['idade_media_socios'] > 0]
```

```python
market = market[market['vl_faturamento_estimado_aux'] < 10000000000]
```

### Corte de features que sobraram com mais de 50% de valores missing

```python
features = aux[aux.percentual > 0.5].index
len(features)
```

```python
market = market[features]
market.drop('qt_socios_st_regular',1, inplace=True)
market.drop(['qt_socios_pj','qt_socios_pf'], 1, inplace=True)
```

```python
market.dropna(inplace=True)
```

```python
market.to_csv('features.csv')
```

### Categ√≥ricas

```python
market.loc[market['fl_rm'] == 'NAO', 'fl_rm'] = 0
market.loc[market['fl_rm'] == 'SIM', 'fl_rm'] = 1
market['fl_rm'] = pd.to_numeric(market['fl_rm'])
```

```python
col_bool = market.dtypes[market.dtypes == 'bool'].index
for col in col_bool:
        market[col] = market[col].astype(int)
```

```python
cat_cols = market.select_dtypes('object').columns
cat_cols = cat_cols[1:]

market_dummies = pd.get_dummies(market, columns=cat_cols, drop_first=True)
```

### To CSV

```python
market_dummies.index = market_dummies.id
market_dummies.drop('id',1,inplace=True)
```

```python
market_dummies.to_csv('pp_market.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
from sklearn.preprocessing import MaxAbsScaler

from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

pd.set_option('display.max_columns', 999)
pd.set_option('display.max_rows', 999)
```

```python
csv_path = '../Data/'

market = pd.read_csv(f'{csv_path}estaticos_market.csv', index_col='Unnamed: 0')
port1 = pd.read_csv(f'{csv_path}estaticos_portfolio1.csv', index_col='Unnamed: 0')
port2 = pd.read_csv(f'{csv_path}estaticos_portfolio2.csv', index_col='Unnamed: 0')
port3 = pd.read_csv(f'{csv_path}estaticos_portfolio3.csv', index_col='Unnamed: 0')

port2 = port2.merge(market,on='id')
port3 = port3.merge(market,on='id')

port1.name  = 'Portfólio 1'
port2.name  = 'Portfólio 2'
port3.name  = 'Portfólio 3'
```

```python
def faturamento_x_funcionarios(df):
    with sns.color_palette("bright", 10):
        fig, ax = plt.subplots(ncols=3, figsize = (30,7))

        sns.scatterplot(data=df, x='vl_faturamento_estimado_aux', y='qt_funcionarios',
                        hue='setor', alpha=0.8, ax = ax[0])

        sns.scatterplot(data=df[df['vl_faturamento_estimado_aux'] < 1e9],
                        x='vl_faturamento_estimado_aux', y='qt_funcionarios',
                        hue='setor', alpha=0.8, ax = ax[1])

        sns.scatterplot(data=df[df['vl_faturamento_estimado_aux'] < 2e8],
                        x='vl_faturamento_estimado_aux', y='qt_funcionarios',
                        hue='setor', alpha=0.8, ax = ax[2])

        ax[0].set_title(f'{df.name} - Faturamento x Funcionários',fontsize=15)
        ax[1].set_title(f'{df.name} - Faturamento x Funcionários (Faturamento < 1.000.000)',fontsize=15)
        ax[2].set_title(f'{df.name} - Faturamento x Funcionários (Faturamento < 200.000)',fontsize=15)

        sns.color_palette("bright", 10)
        ax[0].grid()
        ax[1].grid()
        ax[2].grid()
        plt.show()
        
def quant_funcionarios(df,color):
    plt.figure(figsize=(10,5))
    sns.boxplot(df.loc[df['qt_funcionarios'] < 100, 'qt_funcionarios'], color=color)
    plt.title(f'{df.name} - Boxplot da quantidade de funcionários',fontsize=15)
    plt.xlim(0,100)
    plt.grid()
    plt.show()
```

```python
portfolios = ['Portfólio 1','Portfólio 2','Portfólio 3']
```

# Idade da empresa

```python
plt.figure(figsize=(22,5))
sns.distplot(port1['idade_empresa_anos'],bins=58, color = 'darkturquoise', kde=False, norm_hist=True)
sns.distplot(port2['idade_empresa_anos'],bins=58, color = 'orchid', kde=False, norm_hist=True)
sns.distplot(port3['idade_empresa_anos'],bins=58, color = 'royalblue', kde=False, norm_hist=True)
plt.grid()
plt.title('Histograma da idade da empresa em anos por Portfólio', fontsize=18)
plt.xlim(0,58)
plt.legend(portfolios)
```

# Média de idade dos sócios

```python
plt.figure(figsize=(22,5))
sns.distplot(port1['idade_media_socios'], color = 'darkturquoise', kde=False, norm_hist=True)
sns.distplot(port2['idade_media_socios'], color = 'orchid', kde=False, norm_hist=True)
sns.distplot(port3['idade_media_socios'], color = 'royalblue', kde=False, norm_hist=True)
plt.grid()
plt.title('Distribuição da média de idade dos sócios por Portfólio', fontsize=18)
plt.legend(portfolios)
plt.xlim(0,90)
plt.show()
```

# Renda per capta do setor censitário

```python
plt.figure(figsize=(20,5))
sns.distplot(port1['empsetorcensitariofaixarendapopulacao'],bins=50, color = 'darkturquoise', kde=False, norm_hist=True)
sns.distplot(port2['empsetorcensitariofaixarendapopulacao'],bins=50, color = 'orchid', kde=False, norm_hist=True)
sns.distplot(port3['empsetorcensitariofaixarendapopulacao'],bins=50, color = 'royalblue', kde=False, norm_hist=True)
plt.grid()
plt.title('Distribuição da renda per capta por setor censitário (< R$ 4.000,00) por Portfólio', fontsize=18)
plt.legend(portfolios)
plt.xlim(0,4000)
plt.show()
```

# Faturamento x Funcionários

```python
with sns.color_palette("bright", 10):
    fig, ax = plt.subplots(ncols=3, figsize = (30,7))

    sns.scatterplot(data=port1, x='vl_faturamento_estimado_aux', y='qt_funcionarios',
                    hue='setor', alpha=0.8, ax = ax[0], palette=['limegreen','blueviolet','blue'])

    sns.scatterplot(data=port1[port1['vl_faturamento_estimado_aux'] < 1e9],
                    x='vl_faturamento_estimado_aux', y='qt_funcionarios',
                    hue='setor', alpha=0.8, ax = ax[1], palette=['limegreen','blueviolet','blue'])

    sns.scatterplot(data=port1[port1['vl_faturamento_estimado_aux'] < 2e8],
                    x='vl_faturamento_estimado_aux', y='qt_funcionarios',
                    hue='setor', alpha=0.8, ax = ax[2], palette=['limegreen','blueviolet','blue'])

    ax[0].set_title(f'{port1.name} - Faturamento x Funcionários',fontsize=15)
    ax[1].set_title(f'{port1.name} - Faturamento x Funcionários (Faturamento < 1.000.000)',fontsize=15)
    ax[2].set_title(f'{port1.name} - Faturamento x Funcionários (Faturamento < 200.000)',fontsize=15)

    sns.color_palette("bright", 10)
    ax[0].grid()
    ax[1].grid()
    ax[2].grid()
    plt.show()
    
faturamento_x_funcionarios(port2)
faturamento_x_funcionarios(port3)
```
